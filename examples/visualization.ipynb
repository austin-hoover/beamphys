```python
import importlib
import sys
sys.path.append('..')

import numpy as np
import matplotlib.pyplot as plt
import proplot as pplt
import psdist.visualization as psv
```

```python
x = np.linspace(-4.0, 4.0, 21)
y = np.exp(-0.5 * x**2)

fig, axs = pplt.subplots(ncols=3, nrows=2, figwidth=5.0, share=False)
kinds = ['line', 'step', 'bar']
axs.format(
    xticks=[], yticks=[], 
    toplabels=[f"kind='{kind}'" for kind in kinds],
    leftlabels=[f"flipxy={s}" for s in [False, True]],
)
for j, kind in enumerate(kinds):
    for fill in [False, True]:
        if kind == 'bar' and not fill:
            continue
        alpha = 1.0
        if fill and kind != 'bar':
            kind = kind + 'filled'
            alpha = 0.2      
        for i, flipxy in enumerate([False, True]):
            psv.plot1d(x, y, ax=axs[i, j], kind=kind, flipxy=flipxy, alpha=alpha)
plt.show()
```
